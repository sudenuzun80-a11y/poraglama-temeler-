```
sifre = input("Şifre gir: ")

if len(sifre) >= 6 and any(c.isdigit() for c in sifre) and any(c.isupper() for c in sifre) and any(not c.isalnum() for c in sifre):
    print("Güçlü şifre")
else:
    print("Zayıf şifre")

#soru51
```

```
def ters_liste_olustur(l):
    return l[::-1]

girdi_liste = [12, 70, 90]
ters_liste = ters_liste_olustur(girdi_liste)
print("Orijinal liste:", girdi_liste)
print("Ters çevrilmiş liste:", ters_liste)
#soru71
```

```
def carpim_tablosu(x):
    i = 1
    print(str(x) + " çarpım tablosu:")
    while i <= 10:
        print(str(x) + " x " + str(i)  + " = " + str(x * i))
        i = i + 1
sayi = 8
carpim_tablosu(sayi)
#soru91
```
